Tahap kedua ialah mendefinisikan fungsi detect_text dimana pada fungsi ini, path dari sebuah gambar yang kita gunakan akan dikenali dan akan dibaca dalam bentuk byte string. metode document_text_detection Google Vision API digunaka untuk mendeteksi teks dalam dokumen atau gambar dengan teks yang terstruktur.Lalu, text_annotations Mengambil hasil OCR dari respon API yang kemudian hasilnya dimasukkan ke dalam list ocr_text.

```python
def detect_text(path):
    """Detects text in the file."""

    client = vision.ImageAnnotatorClient()

    with open(path, "rb") as image_file:
        content = image_file.read()

    image = vision.Image(content=content)

   
    response = client.document_text_detection(image=image)
    texts = response.text_annotations
    ocr_text = []

    for text in texts:
        ocr_text.append(f"\r\n{text.description}")

    if response.error.message:
        raise Exception(
            "{}\nFor more info on error messages, check: "
            "https://cloud.google.com/apis/design/errors".format(response.error.message)
        )
    return ocr_text
```

Mendefinisikan path folder gambar input untuk gambar yang sudah difiltering dengan metode sauvola. Pada tahap ini juga akan didefinisikan list yang berfungsi untuk menyimpan output.

```python
# Path folder gambar
input_folder = "./output_100"

# Dictionary untuk menyimpan hasil deteksi teks
detected_texts_sauvola = {}
```

Dikarenakan gamabr yang digunakan cukup banyak dan jika dilakukan manual akan cukup lama maka akan dilakukan perulangan sebanyak jumlah gambar yang ada di folder. Setiap gambar dalam perulangan tersebut akan dideteksi textnya dan hasilnya akan dimasukkan ke dalam dictionary dengan key berupa nama gambarnya sedangkan valuenya ialah hasil dari ocrnya.

```python
# Proses setiap file gambar dalam folder
for file_name in os.listdir(input_folder):
    if file_name.endswith(('.jpg', '.jpeg', '.png')):  # Filter hanya file gambar
        image_path = os.path.join(input_folder, file_name)
        
        try:
            # Deteksi teks pada gambar
            text = detect_text(image_path)
            detected_texts_sauvola[file_name] = text

            # Mencetak hasil deteksi teks untuk setiap gambar
            print(f"\nDetected text for image: {file_name}")
            for line in text:
                print(line)  # Mencetak setiap baris teks yang terdeteksi

        except Exception as e:
            print(f"Error processing {file_name}: {e}")
            detected_texts_sauvola[file_name] = None  # Jika error, simpan None
```

Kode dibawah ialah contoh isi dari dictionary untuk hasil ocr pada seluruh gambar yang belum difiltering oleh metode sauvola.

```python

print("\nFinal Detected Texts Dictionary:")
print(detected_texts_non_sauvola)
```
